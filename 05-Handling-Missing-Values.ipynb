# Handling Missing Values
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

```python
import pandas as pd
```

```python
df = pd.read_csv('wildfires_clean.csv', header=0)
# summarize the dataset
df.head(10)
```

```python
df.info()
```

```python
features = df.columns[1:]
features
```

```python
num_missing = (df[features] == 0).sum()
# report the results
print(num_missing)
```

```python
import numpy as np

df[features] = df[features].replace(0, np.nan)
# count the number of nan values in each column
df.isnull().sum()
```

```python
df.head(20)
```

```python
df.shape
```

```python
df1=df
```

```python
nan_percentages = (df1.isna().sum(axis=1) / len(df1.columns)) * 100
nan_percentages
```

```python
#df.dropna(inplace=True)
#(3507, 43)


threshold = 50
df1.drop(nan_percentages[nan_percentages > threshold].index, inplace=True)
df1.shape
```
