# Handling Missing Values
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

```python
import pandas as pd
```

```python
df = pd.read_csv('pima-indians-diabetes.csv', header=0)
# summarize the dataset
df.head(10)
```

```python
df.describe()
```

```python
features = df.columns[1:6]
features
```

```python
num_missing = (df[features] == 0).sum()
# report the results
print(num_missing)
```

```python
import numpy as np

df[features] = df[features].replace(0, np.nan)
# count the number of nan values in each column
df.isnull().sum()
```
